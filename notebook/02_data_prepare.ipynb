```python
import pandas as pd
```

```python
filename = "../data/volvocars_salesvolumes.csv"
```

```python
df = pd.read_csv(filename, skiprows=2)
```

```python
df.dropna(inplace=True)
df.drop(['Unnamed: 1'], axis=1, inplace=True)
df.set_index('RETAIL SALES BY GEOGRAPHICAL REGIONS', inplace=True)
df = df.transpose()
df.index.name = 'Quarter'
df = df.replace('%', '', regex=True).replace(' - ', None, regex=True).replace(' ', '', regex=True).replace('', None, regex=True)
df = df.astype(float)
```
